```python
#Now we are going on machine learning model and are aim to predict tomorrow max temp
#so we are creating target which we need to predict
core_weather["target"] = core_weather.shift(-1)["temp_max"]
```

```python
core_weather
```

```python
#as we see we have missing value in last row so we are going to get rid of that row
core_weather = core_weather.iloc[:-1,:]
```

```python
core_weather
```

```python
from sklearn.linear_model import Ridge
#we are using Ridge Regression which is a type of regression which minimize overfitting

reg = Ridge(alpha=.1)
```

```python
#we made a variable predictor to predict tomorrow max temp by using precip, temp_max and temp_min
predictors = ["precip","temp_max","temp_min"]
```

```python
#Now we are splitting are data into training and testing set
train = core_weather.loc[:"2022-12-31"]
test = core_weather.loc["2023-01-01":]
#so here it means are traing set is everything before 2022 and testing set is after 2023
```

```python
#now we are fitting this model to regression and making some predictions
reg.fit(train[predictors],train["target"])
```

```python
#we are generating predictions on are test dataset using predictors
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
#we are going to use mean absolute error it basically subtract the actual from the predictions and take the absolute value
#and then find the avg across all the predictions
```

```python
mean_absolute_error(predictions,test["target"])
#the below no is the means on an avg we are 6.2 degree off the actual temp
```

```python
#now we are going to combine the actual value and the predicted value
combined = pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
combined.columns = ["actual","predictions"]
```

```python
#Now we are seeing the actual value and the predictions
combined
```

```python
#now we are plotting to see more clearly
combined.plot()
```

```python
#now we are seeing the cofficient of are regression model how different they look
reg.coef_
#so below values means precipt has negative impact, temp max has big input as we are finding other day temp max and
#then the min temp is also smaller input
```

```python
#now we are creating a function to wrap up everything ahich we done till now
#so splitting the training and test set then generating the predictions and then combining the predictions together
def create_predictions(predictors,core_weather,reg):
  train = core_weather.loc[:"2022-12-31"]
  test = core_weather.loc["2023-01-01":]
  reg.fit(train[predictors],train["target"])
  predictions = reg.predict(test[predictors])
  error = mean_absolute_error(test["target"],predictions)
  combined = pd.concat([test["target"],pd.Series(predictions,index=test.index)],axis=1)
  combined.columns = ["actual","predictions"]
  return error,combined
```

```python
#we are creating more predictors
#what was the avg temp of this month
#we are using pandas rolling method to find certain period rolling mean
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
#so it means it will look back on 30 days and find avg of certain column during that period
```

```python
core_weather
#below we had nan bcz we didnt had 30 days before data above them
```

```python
#Now we are finding some interesting ratio
#so we are looking on a given day was the temp different from the montly mean or not

core_weather["month_day_max"] = core_weather["month_max"]/core_weather["temp_max"]
# Handle potential division by zero or very small values
core_weather.replace([np.inf, -np.inf], np.nan, inplace=True)
core_weather["month_day_max"].fillna(core_weather["month_day_max"].mean(), inplace=True)
```

```python
#we can also look the ratio between the max and the min temp
core_weather["max_min"] = core_weather["temp_max"]/core_weather["temp_min"]
# Handle potential division by zero or very small values
core_weather.replace([np.inf, -np.inf], np.nan, inplace=True)
core_weather["max_min"].fillna(core_weather["max_min"].mean(), inplace=True)
```

```python
#now we are going to update are predictors
predictors = ["precip","temp_max","temp_min","month_max","month_day_max","max_min"]
```

```python
#Now before that we need to remove first 30 rows as they have nan value
core_weather = core_weather.iloc[30:,:].copy()
```

```python
#Now we can create predictions
error, combined = create_predictions(predictors,core_weather,reg)
```

```python
error
#so we can see we have reduced the error
```

```python
#now we are going to see are combined dataframe
combined.plot()
```

```python
#now we are adding monthly and daily avg
core_weather["monthly_avg"] = core_weather["temp_max"].groupby(core_weather.index.month).transform(lambda x: x.expanding(1).mean())
```

```python
#daily avg
core_weather["day_of_year_avg"] = core_weather["temp_max"].groupby(core_weather.index.day_of_year).transform(lambda x: x.expanding(1).mean())
```
